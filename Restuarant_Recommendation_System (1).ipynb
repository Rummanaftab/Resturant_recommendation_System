### df.describe() generates descriptive statistics of the data within the DataFrame df.  It calculates various summary statistics for each numerical column in the DataFrame, such as count, mean, standard deviation, minimum, maximum, and quartiles.

### Count: The count represents the number of non-null (non-missing) values in each numerical column. This statistic provides an indication of the completeness of the data. If the count for a column is less than the total number of rows in the DataFrame, it suggests missing values in that column.

### Mean (Average): The mean is the arithmetic average of all the values in a column. It gives a measure of the central tendency of the data. However, it can be sensitive to outliers, especially in skewed distributions.

### Standard Deviation:The standard deviation measures the dispersion or spread of the values around the mean. A high standard deviation indicates that the values in the column are spread out over a wide range from the mean, while a low standard deviation suggests that the values are clustered closely around the mean.
### 25th Percentile (First Quartile):The 25th percentile, also known as the first quartile (Q1), divides the data into four equal parts, with 25% of the data falling below this value. It gives insight into the spread of the lower values in the data distribution.

### Median (50th Percentile, Second Quartile):The median is the middle value of the sorted column data.It represents the value below which 50% of the data falls and above which the other 50% falls. It's also the second quartile (Q2). Median is often preferred over mean when dealing with skewed distributions or datasets containing outliers because it's less affected by extreme values.
### 75th Percentile (Third Quartile): The 75th percentile, or third quartile (Q3), divides the data into four equal parts, with 75% of the data falling below this value. It provides insight into the spread of the higher values in the data distribution.

### The quantile, specifically the 90th percentile, is a value below which 90% of the data points fall. In other words, 90% of the data points are less than or equal to this value.
### It's a measure of statistical dispersion that indicates the point below which a given fraction of observations in a group of observations fall.


```python
# Example dataset
data = [20, 10, 30, 50, 40]

# Sort the dataset
sorted_data = sorted(data)
print("Sorted data:", sorted_data)
```

```python
# Calculate the index corresponding to the 90th percentile
index_90th_percentile = int(0.9 * (len(sorted_data) - 1))
print("Index for the 90th percentile:", index_90th_percentile)
```

```python
# Retrieve the value at the calculated index
p90_nearest_rank = sorted_data[index_90th_percentile]
print("90th percentile (nearest rank interpolation):", p90_nearest_rank)
```

### Bayesian average to compute average rating, properly
ref: https://en.wikipedia.org/wiki/Bayesian_average

image.png

```python
def weighted_rating(x,m=m,C=C):
    v=x["Votes"]
    R=x["Aggregate rating"]
    return (v/(v+m)*R) + (m/(m+v)*C)
```

### df.apply(): The apply() function in pandas is a powerful method used to apply a function along an axis of a DataFrame.

### df.apply(func, axis=0, raw=False, result_type=None, args=(), **kwds)
### func: The function to apply to each element or row/column of the DataFrame
### axis: The axis along which the function will be applied. Axis 0 applies the function to each column, while axis 1 applies it to each row.
### raw: If True, the passed function will receive ndarray objects instead of Series objects.
### result_type: Specifies the type of the result. It can be 'expand', 'reduce', or 'broadcast'.
### args: Additional arguments to pass to the function.

```python
q_restuarant["score"]=q_restuarant.apply(weighted_rating,axis=1)
```
